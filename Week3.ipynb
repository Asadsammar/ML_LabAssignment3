```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Load dataset
df = pd.read_csv('./insurance.csv')
```

```python
# Define independent and target variables
X = df[['age', 'bmi', 'children']]
y = df['charges']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create linear regression object
reg = LinearRegression()
```

```python
# Train the model
reg.fit(X_train, y_train)
```

```python
# Make predictions on test set
y_pred = reg.predict(X_test)
```

```python
# Model evaluation metrics
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print("R-squared: ", r2)
print("MSE: ", mse)
print("MAE: ", mae)
```

**Explanation:**

The dataset contains information about personal medical costs and various demographic features like age, BMI, number of children etc. The target variable is 'charges' which indicates personal medical costs. The independent variables used are 'age', 'bmi', and 'children'.

The data is split into 80% training and 20% testing sets. Feature scaling is done using StandardScaler to normalize the independent variables. A multiple linear regression model is trained on the training data and predictions are made on the test data.

The model is evaluated using R-squared, Mean Squared Error (MSE) and Mean Absolute Error (MAE).
